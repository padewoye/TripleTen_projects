```python
## Sprint 7 - (A)

import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier     
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
## Sprint 7 - (B-1, B-2)

df = pd.read_csv('/datasets/users_behavior.csv')
df.info()
```

```python
df.head(10)
```

```python
## Sprint 7 - (B-3)

df.isna().sum()
```

```python
## Sprint 7 - (B-4)

df_inf = np.isinf(df)
df_inf_counts = df_inf.sum(axis=0)
print(df_inf_counts)
```

```python
## Sprint 7 - (B-5)

## No need to fill, no NULL or INF values found (see B-3, B-4)
```

```python
## Sprint 7 - (C-1) 
##  Split the data into a TRAINING, VALIDATION and TEST set (3:1:1 Model - 60%, 20% 20%)

##  First split:  divide data into 60% train and 40% temp (this will become validation and test sets)
df_train, df_temp = train_test_split(df, test_size=0.4, random_state=77)

##  Second split:  divide df_temp into df_valid and df_test into 20% validation and 20% test sets
df_valid, df_test = train_test_split(df_temp, test_size=0.5, random_state=77)
```

```python
## Sprint 7 - (C-2) 
##   DF.SHAPE() check to verify that data split correctly into TRAIN, VALID and TEST set (3:1:1 Model - 60%, 20% 20%)

print('df_train shape:', df_train.shape)
print()
print('df_temp shape:', df_temp.shape)
print()
print('df_valid shape:', df_valid.shape)
print()
print('df_test shape:', df_test.shape)
```

```python
## Sprint 7 - (C-3, C-4) 
##   Create the FEATURES and TARGET variables from TRAIN, VALID and TEST datasets
##     SHAPE() validation of the FEATURES and TARGET variables

df_train_features = df_train.drop(['is_ultra'], axis=1)
df_train_target = df_train['is_ultra']

df_valid_features = df_valid.drop(['is_ultra'], axis=1)
df_valid_target = df_valid['is_ultra']

df_test_features = df_test.drop(['is_ultra'], axis=1)
df_test_target = df_test['is_ultra']

print('df_train_features:', df_train_features.shape)
print('df_train_target:', df_train_target.shape)
print()
print('df_valid_features:', df_valid_features.shape)
print('df_valid_target:', df_valid_target.shape)
print()
print('df_test_features:', df_test_features.shape)
print('df_test_target:', df_test_target.shape)
```

```python
df_train_features.head(10)
```

```python
## Sprint 7 - (C-5, C-6, D-1) 
##   Fit and train the model(s)
##      (a) DecisionTreeClassifier

##  Create, Train and Predict a model
##  Create a loop for max_depth from 1 to 9
for depth in range(1, 9):
    model = DecisionTreeClassifier(random_state=77, max_depth=depth)
    
    ##  Train the model
    model.fit(df_train_features, df_train_target)
    
    ##  Find Predictions using the VALID dataset
    predictions_valid = model.predict(df_valid_features)
    
    print("max_depth =", depth, ": ", end="")
    print(accuracy_score(df_valid_target, predictions_valid))
```

```python
## Sprint 7 - (C-5, C-6, D-1)
##   Fit and train the model(s)
##      (b) RandomForestClassifier

##  Create a hyperparameter range from 1 to 30
##  Create, Train and Predict a model
best_score = 0
best_est = 0
for est in range(1, 30):
    ##  set number of trees
    model = RandomForestClassifier(random_state=77, n_estimators=est)
    
    ##  Train on TRAIN model
    model.fit(df_train_features, df_train_target)
    
    ##  Create an accuracy score using the VALID set
    score = model.score(df_valid_features, df_valid_target)
    
    ##  Find BEST score. Display number of estimators it takes to get the BEST score.
    if score > best_score:
        best_score = score
        best_est = est
        
##  Print the BEST score model
print("Using Range 1 - 30")
print("Accuracy of the BEST model, using VALID set (n_estimators = {}):  {}".format(best_est, best_score))
```

```python
## Sprint 7 - (C-5, C-6, D-1)
##   Fit and train the model(s)
##      (b) RandomForestClassifier

##  Create a hyperparameter range from 1 to 50
##  Create, Train and Predict a model
best_score = 0
best_est = 0
for est in range(1, 50):
    ##  set number of trees
    model = RandomForestClassifier(random_state=77, n_estimators=est)
    
    ##  Train on TRAIN model
    model.fit(df_train_features, df_train_target)
    
    ##  Create an accuracy score using the VALID set
    score = model.score(df_valid_features, df_valid_target)
    
    ##  Find BEST score. Display number of estimators it takes to get the BEST score.
    if score > best_score:
        best_score = score
        best_est = est
        
##  Print the BEST score model
print("Using Range 1 - 50")
print("Accuracy of the BEST model, using VALID set (n_estimators = {}):  {}".format(best_est, best_score))
```

```python
## Sprint 7 - (C-5, C-6, D-1)
##   Fit and train the model(s)
##      (b) RandomForestClassifier

##  Create a hyperparameter range from 1 to 120
##  Create, Train and Predict a model
best_score = 0
best_est = 0
for est in range(1, 120):
    ##  set number of trees
    model = RandomForestClassifier(random_state=77, n_estimators=est)
    
    ##  Train on TRAIN model
    model.fit(df_train_features, df_train_target)
    
    ##  Create an accuracy score using the VALID set
    score = model.score(df_valid_features, df_valid_target)
    
    ##  Find BEST score. Display number of estimators it takes to get the BEST score.
    if score > best_score:
        best_score = score
        best_est = est
        
##  Print the BEST score model
print("Using Range 1 - 120")
print("Accuracy of the BEST model, using VALID set (n_estimators = {}):  {}".format(best_est, best_score))
```

```python
## Sprint 7 - (C-5, C-6, D-1)
##   Fit and train the model(s)
##      (c) LogisticRegression

## Initialize the model
model = LogisticRegression(random_state=77, solver='liblinear')

## Train model using the TRAIN set
model.fit(df_train_features, df_train_target)

## Calculate ACCURACY SCORE using the TRAIN set
score_train = model.score(df_train_features, df_train_target)

## Calculate ACCURACY SCORE using the VALID set
score_valid = model.score(df_valid_features, df_valid_target)

##  Print SCORES
print("LogisticRegression SCORE for TRAIN set:", score_train)
print("LogisticRegression SCORE for VALID set:", score_valid)
```

```python
## Sprint 7 - (C-5, D-2)
##   Fit and train the FINAL model
##      (b) RandomForestClassifier

## Use the BEST ESTIMATOR from previous tested models
final_model = RandomForestClassifier(random_state=77, n_estimators=best_est)
    
##  Train on both the TRAIN and VALID data sets
final_model.fit(pd.concat([df_train_features, df_valid_features]), pd.concat([df_train_target, df_valid_target]))

##  Find Predictions using the DF_TEST_FEATURES dataset
predictions_test = final_model.predict(df_test_features)
    
accuracy_test = accuracy_score(df_test_target, predictions_test)

print("Accuracy Score of the FINAL model, using TEST set (best estimator = {}):  {}".format(best_est, accuracy_test))
```
